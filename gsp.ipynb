```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline
```

```python
# Read the data
gsp_df = pd.read_csv("productivity.csv")

#Preprocess data
gsp_df = gsp_df.dropna() # Drops all rows that contain null values
gsp_df = gsp_df.drop_duplicates(keep='first') # Drop duplicate rows if any
gsp_df.info() # Gets info of the existing columns
```

```python
# Creating dataframe with only Control/Treatement data and their dummy/interaction variables 
gsp_df2 = gsp_df.loc[(gsp_df['STATE'] == "NORTH_DAKOTA") | (gsp_df['STATE'] == "SOUTH_DAKOTA")]
gsp_df2['YR79'] = np.where(gsp_df2['YR'] >= 1979, 1, 0)
gsp_df2['TREATMENT'] = np.where(gsp_df2['STATE'] == "SOUTH_DAKOTA", 1, 0)
gsp_df2['YR79_TREATMENT'] = gsp_df2.apply(lambda row: (row['YR79']*row['TREATMENT']), axis=1)
gsp_df2 = gsp_df2.reset_index(drop=True)
gsp_df2.head()
```

```python
# First Difference Estimate
gsp_df3 = gsp_df2.loc[gsp_df2['STATE'] == "SOUTH_DAKOTA"]
result = sm.OLS(gsp_df3['GSP'], sm.add_constant(gsp_df3[['YR79']])).fit()
result.summary()
```

```python
# Comparing North Dakota and South Dakota
D1 = gsp_df2.loc[gsp_df2['STATE'] == "NORTH_DAKOTA"].describe()
D2 = gsp_df2.loc[gsp_df2['STATE'] == "SOUTH_DAKOTA"].describe()
D3 = D1 - D2
D3
```

```python
# Difference in Differences
result = sm.OLS(gsp_df2['GSP'], sm.add_constant(gsp_df2[['YR79', 'TREATMENT', 'YR79_TREATMENT']])).fit()
result.summary()
```

```python
# Verifying the DD model
a = gsp_df2.loc[(gsp_df2['YR79'] == 0) & (gsp_df2['TREATMENT'] == 0)].mean()
print ("a : ", a.GSP)

b = gsp_df2.loc[(gsp_df2['YR79'] == 0) & (gsp_df2['TREATMENT'] == 1)].mean()
print ("b : ", b.GSP)

c = gsp_df2.loc[(gsp_df2['YR79'] == 1) & (gsp_df2['TREATMENT'] == 0)].mean()
print ("c : ", c.GSP)

d = gsp_df2.loc[(gsp_df2['YR79'] == 1) & (gsp_df2['TREATMENT'] == 1)].mean()
print ("d : ", d.GSP)

estimate = (d.GSP-c.GSP) - (b.GSP-a.GSP)
print ("Difference in Differnece estimate : ", estimate)
```
